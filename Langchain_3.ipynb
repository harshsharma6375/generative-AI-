```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template("what is a good name for a company that makes {product}")
prompt.format(product = "colourful socks")
```

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt | llm | StrOutputParser()

# invoke the chain
response = chain.invoke({"product": "colorful socks"})

# print the response
print(response)
```

```python
# multiple chaining
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence

# STEP 3 - INTIALIZE LLM
llm = ChatGoogleGenerativeAI(model = "gemini-2.0-flash",temperature=0.9)

# STEP 4 - FIRST PROMPT : RESTAURANT NAME
prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="I want to open a restaurant for {cuisine} food. Suggest a fancy name for this."
)

# STEP 5 - SECOND PROMPT : MENU ITEM
prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}. Return it as a comma separated string"
)

# STEP 6 - dEFINE THE TWO LLM PIPELINES
name_chain = prompt_template_name | llm | StrOutputParser()
food_items_chain = prompt_template_items | llm | StrOutputParser()

# STEP 7 - COMBINE BOTH IN SEQUENCES
from langchain_core.runnables import RunnablePassthrough
from langchain_core.runnables import RunnableLambda

# create sequential chain:name -> menu
chain = name_chain | RunnableLambda(lambda name : {"restaurant_name" : name }) | food_items_chain

# invoke the chain
response = chain.invoke({"cuisine" : "Indian"})

# print the response
print(response)

# STEP 8 -  RUN THE PIPELINE
content = chain.invoke({"cuisine" : "Indian"})
print("output:", content)
```
