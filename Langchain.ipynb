```python
# ğŸ“˜ Install if running on Google Colab (optional)
# !pip install json

import json

# ---------------------------
# Sample JSON (can replace with your own)
# ---------------------------
data = {
    "user": "Papa",
    "role": "Data Scientist",
    "skills": ["Python", "AI", "Machine Learning"],
    "project": {
        "name": "GenAI Automation",
        "type": "LLM Pipeline"
    }
}

# ---------------------------
# Step 1: Create Token Dictionary
# ---------------------------
token_dict = {}
reverse_dict = {}
token_counter = 100

def get_token(value):
    global token_counter
    if value not in token_dict:
        token_dict[value] = token_counter
        reverse_dict[token_counter] = value
        token_counter += 1
    return token_dict[value]

# ---------------------------
# Step 2: Convert JSON â†’ TOON (Recursive)
# ---------------------------
def json_to_toon(obj):
    if isinstance(obj, dict):
        return {k: json_to_toon(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [json_to_toon(i) for i in obj]
    elif isinstance(obj, str):
        return get_token(obj)
    else:
        return obj

toon_data = json_to_toon(data)

# ---------------------------
# Step 3: Display Results
# ---------------------------
print("ğŸ”· Original JSON:")
print(json.dumps(data, indent=2))

print("\nğŸ”¶ TOON Format:")
print(json.dumps(toon_data, indent=2))

print("\nğŸ”‘ Token Dictionary:")
print(json.dumps(reverse_dict, indent=2))
```
